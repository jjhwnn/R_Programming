```r
# 구분(사망자수, 발생건수, 부상자수) 컬럼화 / \1~12월 컬럼 로우데이터로 변경

acc3 <- data.frame(발생건수=0, 사망자수=0, 부상자수=0)

for (i in seq(4, nrow(acc2), 3)){

    acc3 <- rbind(acc3 ,t(acc2[c(i:i+2),])
}

acc3
```
