### 기본통계량으로 분석해 보기

```r
# 밀가루 사용

flour <- c(3, -2, -1, 0, -2)

# 다이어트 약
diet <- c(-4, 1, -3, -5, -2, -8)
```

```r
total <- c(flour, diet)
total
```

```r
# 기본 통계량

summary(total)
```

```r
# 히스토그램
hist(total)
```

```r
boxplot(total)
```

```r
boxplot(flour, diet)
```

```r
cor(flour, diet)
```

```r
x <- c(1, 3, 9)
y <- c(10, 20, 30)
```

```r
par(mfrow = c(1,2))
plot(x, type='l')
plot(y, type="l")
```

```r
cor(x,y)
```

```r
cafe <- read.csv("../Data//coffee.csv")
head(cafe)
```

```r
# 커피의 빈도수

table(cafe$Coffee)
```

```r
# 범위 빈도수

coffee_Range <- seq(0, 60, 10) # 범위
table(cut(cafe$Coffee, breaks = coffee_Range, right=F)) # 끝값을 다음에 넘겨줌
```

```r
# 줄기-잎 그림

stem(cafe$Coffee)
```

```r
# 커피와 주스의 변동계수(표준편차sd/평균mean), 표준지수(m/sd)

head(cafe)

cofCv <- sd(cafe$Coffee)  / mean(cafe$Coffee)
cofCv

juiceCv <- sd(cafe$Fruit.Cup) / mean(cafe$Fruit.Cup)
juiceCv
```

```r
sort(cafe$Fruit.Cup)
```

```r
sort(cafe$Coffee)
```
