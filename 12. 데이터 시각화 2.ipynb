### 한 화면에 차트 여러개 그려서 비교하기

```r
price <- runif(10, 1, 100)
plot(price)
```

```r
# 2행 2열 차트 그리기 : type 속성으로 구분해 보기

par(mfrow = c(2,2))
plot(price, type='l')
plot(price)
plot(price, type='h')
plot(price, type='s')
```

```r
# 2행 2열 차트 그리기 : pch 속성으로 구분해 보기
# 어떤 모양으로 사용하는게 좋을지 비교해볼때 사용

par(mfrow = c(2,2))

plot(price, type="o", pch=5)
plot(price, type="o", pch=15)
plot(price, type="o", pch=20, col='blue')
plot(price, type="o", pch=c(rep(15,8),20, 20)
                     , col='orange',
                       cex=c(rep(1,8),3,1)               
    )
```

```r
# 투표수가 높은 시군구의 후보별 득표수의 막대그래프에 시군구별 투표수 표시하기

election <- read.csv("../Data/election_2017.csv")
head(election)
```

```r
library(dplyr)
options(warn = -1)
top5 <- head(arrange(election, desc(pop)),5)[,c(2,3,5:7)]
top5
sum <-top5$moon+top5$hong+top5$ahn
```

```r
top5

par(family = "AppleGothic")
options(scipen = 100)
barplot(
    t(top5[,3:5]),
    main = "투표수 높은 진역의 후보별 투표수",
    names = paste(top5$광역시도, top5$시군),
    ylim = c(0, 500000),
    # xlab = paste( ),
    col = c('skyblue','red','lightgreen'),
    legend = T,
    cex.names = 0.6
)
for(i in 1:nrow(top5)){
    text(0.7+1.2*(i-1), sum[i]+10000, sum[i])
}



```

```r
par(family = "AppleGothic")
options(scipen = 100)
barplot(
    t(top5[,3:5]),
    main = "투표수 높은 진역의 후보별 투표수",
    names = paste(top5$광역시도, top5$시군),
    ylim = c(0, 500000),
    # xlab = paste( ),
    col = c('skyblue','red','lightgreen'),
    legend = T,
    cex.names = 0.6
)

round(top5$moon / sum * 100, 2)


for(i in 1:nrow(top5)){
    sum <-top5$moon[i]+top5$hong[i]+top5$ahn[i]
    text(0.7+1.2*(i-1), top5[i,3]/2, paste(round(top5[i,3] / sum * 100, 2), "%" ))
    text(0.7+1.2*(i-1), top5[i,4]/2+top5[i,3], paste(round(top5[i,4] / sum * 100, 2), "%" ))
    text(0.7+1.2*(i-1), top5[i,5]/2+top5[i,3]+top5[i,4], paste(round(top5[i,5]/ sum * 100, 2), "%" ))
}
```

# 수학 함수의 시각화

```r
f <- function(x){
    y=x^2
    return(y)
}
```

```r
x <- 1:3
f(x)
```

```r
x <- seq(-5, 5, 0.1)

plot(x, f(x),
     type="l"
    )
```
