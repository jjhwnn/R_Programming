# column이름과 one-hot encoding을 한번에 처리하기

```r
mr3 <- data.frame(V1 = mr$V1)
head(mr3)
```

```r
count <- 2 #

for(i in 2:23){
    for(j in sort(unique(mr[,i]))){
        
        mr3 <- cbind(mr3, mr[,i] == j)
        colnames(mr3)[count] <- paste("V", i, "_", j, sep='')
        count <- count + 1
        
    }
}

head(mr3)
```

```r
mr3 <- arrange(mr3, V1)
```

```r
library(caret)
library(nnet)

set.seed(1234)

samp <- createDataPartition(mr3$V1, p=0.7, list=F)

data.train <- mr3[samp,]
str(data.train)

data.test <- mr3[-samp,]
str(data.test)

x <- subset(data.test, select = -V1)
y <- data.test$V1
head(samp)
```

```r
# 인공신경망

model.nnet <- function(x){
    model.nnet <- nnet(V1 ~ ., data=data.train, size=3) 

    pred <- predict(model.nnet, x, type = "class")

    per <- mean(pred == y)
    
    return (per)
}

model.nnet(x)
```

```r
# 의사결정나무
library(party)

model.ctree <- function(x){
    model.ctree <- ctree(V1 ~ ., data=data.train) 

    pred <- predict(model.ctree, x)

    per <- mean(pred == y)
    
    return (per)
}

model.ctree(x)
```

```r
# randomForest
library(randomForest)

rf <- function(x){

    
    model.rf <- randomForest(V1 ~ ., data = data.train, ntree = 100)
    
    pred3 <- predict(model.rf, x)

    mean(pred3 == y)
    
}

rf(x)



```

```r
library(e1071)
ssvm <- function(x) {

    model.svm <- svm(V1 ~ ., data=data.train)

    pred4 <- predict(model.svm, x, type="class")

    mean(pred4 == y)
}

ssvm(x)
```

```r
# 다중 로지스틱 회귀분석

model <- function(x) {
    
    model <- multinom(V1 ~., data=data.train)
                  
    pred5 <- predict(model, x, type="class")

    mean(pred5 == y)
}

model(x)
```
